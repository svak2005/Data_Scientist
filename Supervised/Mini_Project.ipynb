```python
import matplotlib.pyplot as plt
import pandas as pd
student_info=pd.read_csv('/content/student_performance_dataset_50.csv')
student_info
```

```python
input_dataset=student_info.drop(columns=['InternetAtHome','FinalResult'])
input_dataset
```

```python
output_dataset=student_info[['InternetAtHome','FinalResult']]
output_dataset
```

```python
from sklearn.tree import DecisionTreeClassifier,plot_tree
student_model=DecisionTreeClassifier()
student_model.fit(input_dataset,output_dataset)
new_student = pd.DataFrame([[2.0, 5, 1000, 10, 60, 12]], columns=input_dataset.columns)
prediction = student_model.predict(new_student)
print("Prediction:", prediction)
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
plot_tree(
    student_model,
    feature_names=input_dataset.columns,
    class_names=[
        list(output_dataset['InternetAtHome'].unique().astype(str)) +
        list(output_dataset['FinalResult'].unique().astype(str))
    ],
    filled=True,
    rounded=True,
    ax=ax
)
plt.title("Multi-output Decision Tree for Student Performance")
plt.show()
```
